```python
from ultralytics import YOLO
import cv2
import time

# Load YOLO model
model = YOLO('yolov8n.pt')

# Open webcam
cap = cv2.VideoCapture(0)  # 0 for default webcam, change if necessary

# Get webcam properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Time variables
left_side_time = 0
right_side_time = 0
prev_time = time.time()

while True:
    # Read frame from webcam
    ret, frame = cap.read()
    if not ret:
        break

    # Detect objects
    results = model(frame)

    # Check if any objects are detected
    if results.pred[0] is not None:
        # Assuming the first detected object is the one you want to track
        x_min, _, width, _ = results.pred[0][0]  # Assuming YOLOv5 output format [x_min, y_min, width, height]
        center_x = x_min + width / 2

        # Update time spent on each side based on the object's position
        if center_x < frame_width / 2:
            left_side_time += time.time() - prev_time
        else:
            right_side_time += time.time() - prev_time

        prev_time = time.time()

    # Display frame
    cv2.imshow('frame', frame)

    # Break loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()

# Print the time spent on each side
print("Time spent on the left side:", left_side_time)
print("Time spent on the right side:", right_side_time)
```
